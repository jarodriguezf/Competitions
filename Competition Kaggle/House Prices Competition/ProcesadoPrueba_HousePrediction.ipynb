```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("darkgrid")
from sklearn.impute import SimpleImputer
from scipy.stats import stats
```

```python
pd.set_option('display.max_rows', None)  
pd.set_option('display.max_columns', None)  
```

```python
df = pd.read_csv("test.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Alley','MasVnrType', 'FireplaceQu', 'PoolQC' ,'Fence','MiscFeature'], axis = 1)
```

```python
df.isnull().sum()
```

```python
#Imputar valores categoricos NaN con la moda
def imputeObject (df):
    imp_cat = SimpleImputer(strategy='most_frequent')
    columnas = df.select_dtypes(include=['object'])
    for i in columnas:
        df[[i]] = imp_cat.fit_transform(df[[i]])
    return df

df = imputeObject(df)
```

```python
#imputar mediana de GarageYrBlt
imp_GarageYrBlt = SimpleImputer(strategy='median')
df[['GarageYrBlt']]=imp_GarageYrBlt.fit_transform(df[['GarageYrBlt']])
```

```python
#Eliminar atipicos de LotFrontage
# Calcula el rango intercuartílico (IQR)
Q1 = df['LotFrontage'].quantile(0.25)
Q3 = df['LotFrontage'].quantile(0.75)
IQR = Q3 - Q1

# Define un umbral para identificar valores atípicos
umbral_superior = Q3 + 1.5 * IQR
umbral_inferior = Q1 - 1.5 * IQR

# Filtra los registros que no son atípicos
df = df[(df['LotFrontage'] >= umbral_inferior) & (df['LotFrontage'] <= umbral_superior)]
```

```python
#imputar mediana de LotFrontage
imp_LotFrontage = SimpleImputer(strategy='median')
df[['LotFrontage']]=imp_LotFrontage.fit_transform(df[['LotFrontage']])
```

```python
#Eliminar atipicos de MasVnrArea
# Calcula el rango intercuartílico (IQR)
Q1 = df['MasVnrArea'].quantile(0.25)
Q3 = df['MasVnrArea'].quantile(0.75)
IQR = Q3 - Q1

# Define un umbral para identificar valores atípicos
umbral_superior = Q3 + 1.5 * IQR
umbral_inferior = Q1 - 1.5 * IQR

# Filtra los registros que no son atípicos
df = df[(df['MasVnrArea'] >= umbral_inferior) & (df['MasVnrArea'] <= umbral_superior)]
```

```python
#imputar mediana de MasVnrArea
imp_MasVnrArea = SimpleImputer(strategy='median')
df[['MasVnrArea']]=imp_MasVnrArea.fit_transform(df[['MasVnrArea']])
```

```python
df.isnull().sum()
```

```python
#imputar mediana de MasVnrArea
imp_rest = SimpleImputer(strategy='median')
df[['BsmtFinSF1', 'BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','BsmtFullBath','BsmtHalfBath','GarageCars','GarageArea']]=imp_rest.fit_transform(df[['BsmtFinSF1', 'BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','BsmtFullBath','BsmtHalfBath','GarageCars','GarageArea']])
```

```python
# Eliminar valor de categoria con baja frecuencia
df = df[df['Condition1'] != 'RRNe'].reset_index(drop=True)
df = df[df['Condition2'] != 'PosA'].reset_index(drop=True)
df = df[df['RoofMatl'] != 'Membran RoofMatl'].reset_index(drop=True)
df = df[~df['Exterior1st'].isin(['AsphShn','Stone','ImStucc','CBlock'])].reset_index(drop=True)
df = df[~df['Exterior2nd'].isin(['Other','CBlock'])].reset_index(drop=True)
df = df[df['ExterCond'] != 'Po'].reset_index(drop=True)
df = df[df['Foundation'] != 'Wood'].reset_index(drop=True)
df = df[df['BsmtCond'] != 'Po'].reset_index(drop=True)
df = df[df['HeatingQC'] != 'Po'].reset_index(drop=True)
df = df[df['Electrical'] != 'Mix'].reset_index(drop=True)
```

```python
# Calcula los Z-Scores para todas las columnas numéricas en el DataFrame
z_scores = np.abs(stats.zscore(df.select_dtypes(include='number')))

# Define un umbral (por ejemplo, 2) para considerar los valores atípicos
threshold = 3

# Filtra el DataFrame para eliminar las filas con Z-Scores mayores que el umbral
df_no_outliers = df[(z_scores <= threshold).all(axis=1)]
```
