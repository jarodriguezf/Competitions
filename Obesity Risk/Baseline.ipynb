```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
train_df = pd.read_csv('train.csv', index_col='id')
train_df
```

```python
train_df.info()
```

```python
# Dividimos en X e y (variables predictoras y objetivo)
X = train_df.iloc[:, :-1]
y = train_df.iloc[:, 16]
```

```python
# matriz pandas con valores de variables predictoras
X
```

```python
# array numpy con valores de variable objetivo
y
```

```python
# Codificar variables X categoricas
categorical_X = X.select_dtypes(include='object')

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
categorical_X_encoded = encoder.fit_transform(categorical_X)

categorical_X = pd.DataFrame(categorical_X_encoded, columns=encoder.get_feature_names_out(categorical_X.columns))
categorical_X
```

```python
# Normalizar variables X numericas
numerical_X = X.select_dtypes(exclude='object')
columns_name_num = numerical_X.columns

normalizer = StandardScaler()
numerical_X_normalized = normalizer.fit_transform(numerical_X)

numerical_X = pd.DataFrame(numerical_X_normalized, columns= columns_name_num)
numerical_X
```

```python
# Concatenamos las variables categoricas ya codificadas y las numericas estandarizadas
X_train = pd.concat([categorical_X, numerical_X], axis = 1)
X_train
```

```python
# Separar en validacion y entrenamiento
X_entrenamiento, X_validacion, y_entrenamiento, y_validacion = train_test_split(X_train, y, test_size=0.2,random_state=123, stratify=y)
```

```python
X_entrenamiento.shape
```

```python
y_entrenamiento.shape
```

```python
# Prediccion del modelo base
base_model = LogisticRegression(multi_class='multinomial', verbose=1, n_jobs=-1, random_state=123)
prediction_train = base_model.fit(X_entrenamiento, y_entrenamiento)
```

```python
# modelo base entrenado
prediction_train
```

```python
# Prediccion del conjunto de validacion
y_pred = prediction_train.predict(X_validacion)
```

```python
y_pred
```

```python
# Metrica accuracy score (requerida para la competicion)
print('El modelo base alcanza:',round(accuracy_score(y_validacion, y_pred), 2)*100,'% de precision')
```

```python
# matriz de confusion
confusion_matrix(y_validacion, y_pred)
```
