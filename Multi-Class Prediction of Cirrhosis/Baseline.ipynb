```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import log_loss
```

```python
train_df = pd.read_csv('train.csv')
train_df
```

```python
test_df = pd.read_csv('test.csv')
test_df
```

#### Separamos dataset para primer modelo base

```python
X = train_df.iloc[:, 1:-1]
y = train_df.iloc[:, 19]
```

```python
X.info()
```

```python
y
```

```python
# normalizar
numerical_data= X.select_dtypes(include='number')
columns_names_numerical = numerical_data.columns
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numerical_data)
```

```python
scaled_data # datos numericos escalados
```

```python
# codificar categoricos
categorical_data = X.select_dtypes(exclude='number')
encoder = OneHotEncoder(handle_unknown='ignore', drop='first', sparse_output=False)
encoded_data = encoder.fit_transform(categorical_data)
columns_names_categorical = encoder.get_feature_names_out(categorical_data.columns)
```

```python
encoded_data # matriz densa
```

```python
# Transformo a dataframe, para asignar nombre de columna 
X_numerical = pd.DataFrame(scaled_data, columns=columns_names_numerical)
X_categorical = pd.DataFrame(encoded_data, columns=columns_names_categorical)
# concat de ambas tranformaciones
X_original = pd.concat([X_numerical, X_categorical], axis = 1)
X_original
```

```python
# validacion y entrenamiento
X_train,X_val,y_train,y_val = train_test_split(X_original, y, test_size=0.2, random_state=0, stratify=y, shuffle=True)
```

```python
# modelo predictivo base
lg = LogisticRegression(solver='lbfgs', random_state=0, max_iter=100, multi_class='multinomial')
model = lg.fit(X_train, y_train)
```

```python
# prediccion de probabilidad 
predict_proba = model.predict_proba(X_val)
```

```python
predict_proba # objetivo de la competicion
```

```python
# Metrica
print(log_loss(y_val, predict_proba))
```

#### Prediccion con datos de prueba

```python
test_df.head()
```

```python
id_ = test_df['id']
X_test = test_df.drop('id', axis = 1)
```

```python
#normalizar
numerical_= X_test.select_dtypes(include='number')
columns_names_n = numerical_.columns
scaler = StandardScaler()
scaled_ = scaler.fit_transform(numerical_)
# codificar categoricos
categorical_ = X_test.select_dtypes(exclude='number')
encoder = OneHotEncoder(handle_unknown='ignore', drop='first', sparse_output=False)
encoded_ = encoder.fit_transform(categorical_)
columns_names_c = encoder.get_feature_names_out(categorical_.columns)
```

```python
# Transformo a dataframe, para asignar nombre de columna 
X_num = pd.DataFrame(scaled_, columns=columns_names_n)
X_cat = pd.DataFrame(encoded_, columns=columns_names_c)
# concat de ambas tranformaciones
X_test = pd.concat([X_num, X_cat], axis = 1)
X_test
```

```python
# Prediccion con datos de prueba
predict_proba_test = model.predict_proba(X_test)
```

```python
predict_proba_test
```

```python
# Guardar en listas cada etiqueta con la probabilidad
Status_C = []
Status_CL = []
Status_D = []
for i in range(len(predict_proba_test)):
    Status_C.append(predict_proba_test[i, 0])
    Status_CL.append(predict_proba_test[i, 1])
    Status_D.append(predict_proba_test[i, 2])
```

```python
# Creamos la submission
submission_test = {'id': id_, 'Status_C':Status_C, 'Status_CL':Status_CL,'Status_D':Status_D}
```

```python
data = pd.DataFrame(submission_test)
data
```

```python
#Guardamos
data.to_csv('baseline.csv', index=False)
```
